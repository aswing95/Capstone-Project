## Build Kmeans model after outlier removal and encoding

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test=train_test_split(X,test_size=0.30,random_state=10)
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for k in range(1,8):
    
    # fit K-means clustering with number of clusters K
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X_train)
    
    # use 'inertia_' to get wcss 
    # append the value of WCSS for each K to the list 'wcss'
    wcss.append(kmeans.inertia_)

# print the within cluster sum of squares
print(wcss)
```

```python
# set the plot size using 'figsize'
plt.figure(figsize=(15,8))

# plot the elbow plot
# pass the values for K and WCSS
# last number in the range() is exclusive
plt.plot(range(1,8), wcss) 

# set the axes and plot labels
plt.xlabel("Number of Clusters")
plt.title('Scree Plot for Optimal Number of Clusters')
plt.ylabel("WCSS")

# display the plot
plt.show()
```

```python
from sklearn.metrics import silhouette_score 
from sklearn.metrics import silhouette_samples
```

```python
silhouette_coefficients = []

# use 'for' loop to build the clusters
# pass the range to iterate the value of k
# last number in the range() is exclusive
for k in range(2,8):
    
    # perform K-means with different K values
    kmeans = KMeans(n_clusters=k, random_state=10)
    # fit and predict on the scaled data
    kmeans.fit(X_train)
    
    # 'silhouette_score' function computes the silhouette score for each K
    # 'random_state' returns the same sample each time you run the code  
    score = silhouette_score(X_train,kmeans.labels_, random_state=10)
    
    # append score to 'silhouette_coefficients'
    silhouette_coefficients.append(score)
    
# print the silhouette coefficients
print(silhouette_coefficients)
```

```python
# set the figure size using 'figsize'
plt.figure(figsize=(15,8))

# plot bar plot of silhouette coefficients
# last number in the range() is exclusive
plt.bar(range(2,8), silhouette_coefficients)

# add axes and plot labels
plt.title('Silhouette Coefficients')
plt.xlabel("Number of Clusters")
plt.ylabel("Silhouette Coefficient")

# display the bar plot
plt.show()
```

```python
Kmeans3=KMeans(n_clusters =2, random_state = 15, max_iter = 200, tol = 0.001)

# fit the model on the scaled data
Kmeans3.fit(X_train)
```

```python
result = Kmeans3.labels_

# create a dataframe of cluster labels
# pass the column name to the parameter 'cluster'
result = pd.DataFrame(result, columns = ["cluster"])

# group the data based on cluster labels and get size of each cluster
obs = result.groupby('cluster').size()

# print the cluster size
result
```

```python
mean_vec=np.mean(X,axis=0)
```

```python
cov_mat=(X-mean_vec).T.dot((X-mean_vec))/(X.shape[0]-1)
cov_mat
```

```python
cov_mat=np.cov(X.T)
cov_mat
```

```python
eig_vals,eig_vects=np.linalg.eig(cov_mat)
```

```python
pca.n_components_
```

### Build a K-Means Clustering Pipeline after PCA

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler
preprocessor = Pipeline(
       [
           ("scaler", MinMaxScaler())
       ]
   )
```

```python
clusterer = Pipeline(
   [
   (
    "kmeans",
     KMeans(
               n_clusters=2,
               init="k-means++",
               n_init=50,
               max_iter=500,
               random_state=42,
    ),
   ),
     ]
   )
```

```python
pipe = Pipeline([('scaler',preprocessor), ('cluster', clusterer)])
```

```python
X=pca_df
X_train,X_test=train_test_split(X,test_size=0.30,random_state=13)
pipe.fit(X_train)
```

```python
preprocessed_data = pipe["scaler"].transform(X_test)
preprocessed_data
```

```python
predicted_labels = pipe["cluster"]["kmeans"].predict(preprocessed_data)
predicted_labels
```

```python
silhouette_score(preprocessed_data, predicted_labels)
```

```python
result = pipe["cluster"]["kmeans"].labels_

# create a dataframe of cluster labels
# pass the column name to the parameter 'cluster'
result = pd.DataFrame(result, columns = ["cluster"])

# group the data based on cluster labels and get size of each cluster
obs = result.groupby('cluster').size()

# print the cluster size
obs
```

```python
from kmodes.kprototypes import KPrototypes
```

```python
# Elbow curve to find optimal K
cost = []
K = range(2,30)
for num_clusters in list(K):
    kmode = KPrototypes(n_clusters=num_clusters, init = "random",max_iter=20, n_init = 10, verbose=1)
    kmode.fit_predict(X_train,categorical=[7,8,9,10,11])
    cost.append(kmode.cost_)
    
plt.plot(K, cost, 'bx-')
plt.xlabel('No. of clusters')
plt.ylabel('Cost')
plt.title('Elbow Method For Optimal k')
plt.show()
```
